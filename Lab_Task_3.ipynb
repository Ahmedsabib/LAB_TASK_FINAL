```python
# Task 3
# Apply Image(DIP_Lab_task_03.png) smoothing with appropriate kernel applied. Identify the noise(if any) before processing.

# There is salt and pepper noise in this given image

import numpy as np
import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow

image = cv2.imread('Lab_Task_Resource/DIP_Lab_task_03.png')
# cv2_imshow(image)
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

def smooth_image(image, filter):
  filter_height, filter_width = filter.shape
  padded_hight = filter_height//2
  padded_width = filter_width//2

  padded_image = np.pad(image, ((padded_hight, padded_hight), (padded_width, padded_width)), 'constant')

  x, y = image.shape
  output = np.zeros((x, y))

  for i in range(x):
    for j in range(y):
        roi = padded_image[i : i + filter_height, j : j + filter_width]
        output[i, j] = np.median(roi)

  return output

smoothing_filter = np.zeros((5, 5), dtype=np.uint8)
smooth_image = smooth_image(image_gray, smoothing_filter)

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(image_gray, cmap='gray')
plt.axis('off')
plt.title('Image 1')

plt.subplot(1, 2, 2)
plt.imshow(smooth_image, cmap='gray')
plt.axis('off')
plt.title('Image 2')
```
