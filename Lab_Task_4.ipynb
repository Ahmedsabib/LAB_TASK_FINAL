```python
# Task 4
# Implement an appropriate technique to reduce the brightness of the provided image (DIP_Lab_task_04.jpg) effectively

import numpy as np
import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow

def gamma_transformation(pixel_data, gamma, max_color):
    c = 1

    for i in range(0, len(pixel_data), 3):
        for j in range(3):

            pixel_data[i + j] = int(c * (pixel_data[i + j] / max_color) ** gamma * max_color)
    return pixel_data

image = cv2.imread('Lab_Task_Resource/DIP_Lab_task_04.jpg')
# cv2_imshow(image)

dark_image = gamma_transformation(image.flatten(), 5, 255)
dark_image = dark_image.reshape(image.shape)

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.axis('off')
plt.title('Image 1')

plt.subplot(1, 2, 2)
plt.imshow(dark_image, cmap='gray')
plt.axis('off')
plt.title('Image 2')
```
