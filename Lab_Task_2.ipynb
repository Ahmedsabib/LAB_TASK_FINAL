```python
# Task 2
# Convert the RGB image(DIP_Lab_task_02.jpg) to a gray scale image and compare a sub-region of the image by intensity values(5*5 dimension would be enough for comparison)

import cv2
import matplotlib.pyplot as plt
import numpy as np
from google.colab.patches import cv2_imshow

image = cv2.imread('Lab_Task_Resource/DIP_Lab_task_02.jpg')
# cv2_imshow(image)

def rgb_to_gray_scale(image):
  image_dim = image.shape
  x = image_dim[0]
  y = image_dim[1]
  z = image_dim[2]
  gray_image = np.zeros((x, y), dtype=np.uint8)

  for i in range(x):
    for j in range(y):
      r, g, b = image[i, j]
      gray_image[i, j] = r * 0.29 + g * 0.58 + b * 0.11

  return gray_image

# Convert to gray scale
gray_image = rgb_to_gray_scale(image)

# 5 * 5 dimension sub-region comparison
plt.figure(figsize=(5, 5))

plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.axis('off')
plt.title('Image 1')

plt.subplot(1, 2, 2)
plt.imshow(gray_image, cmap='gray')
plt.axis('off')
plt.title('Image 2')

```
